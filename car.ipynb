 Identify any potential outliers in a numerical column of your choice. Explain
your approach.


````python
q1=np.percentile(df['cylinders'],25,interpolation='midpoint')
q2=np.percentile(df['cylinders'],50,interpolation='midpoint')
q3=np.percentile(df['cylinders'],75,interpolation='midpoint')
````

````python
iqr=q3-q1
````

````python
iqr
````

````python
minval=q1-(1.5*iqr)
maxval=q3+(1.5*iqr)
````

````python
outliers=[]
for x in df['cylinders']:
  if((x>maxval) or (x<minval)):
    outliers.append(x)
outliers
````
